```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import twint
```

```python
# Configure
c = twint.Config()
```

```python
import nest_asyncio

nest_asyncio.apply()
```

```python
c.Search = "bitcoin"
c.Format = "Username: {username} |  Tweet: {tweet}"
c.Limit = 1
c.Pandas = True
c.Output = "none"
c.Lang = "en"
# Run
twint.run.Search(c)
```
